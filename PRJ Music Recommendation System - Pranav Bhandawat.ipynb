```python
## Addition: Popular Songs by Year Range
```

```python
def recommend_popular_by_year(user_id, min_year, max_year):
    # Step 1: Filter songs released between min_year and max_year
    filtered_songs = song_df[(song_df['year'] >= min_year) & (song_df['year'] <= max_year)]

    # Step 2: Initialize and create popularity recommender
    pr_filtered = Recommenders.popularity_recommender_py()
    pr_filtered.create(filtered_songs, 'user_id', 'song')

    # Step 3: Get recommended songs (contains user_id, song, score, rank)
    recommendations = pr_filtered.recommend(user_id)

    # Step 4: Add year info by merging with filtered_songs
    recommendations_with_year = recommendations.merge(
        filtered_songs[['song', 'year']].drop_duplicates(), 
        on='song', 
        how='left'
    )

    # Step 5: Return clean output
    return recommendations_with_year[['user_id', 'song', 'year', 'score']]
```

```python
recommend_popular_by_year(100, 2000, 2020)
```
